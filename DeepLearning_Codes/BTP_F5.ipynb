```python
from PIL import Image
import os
import random

def crop_and_replace_images(folder_path, left, top, right, bottom):
    # Check if the folder path exists
    if not os.path.exists(folder_path):
        print(f"Error: The folder path '{folder_path}' does not exist.")
        return

    # Get a list of all files in the folder
    image_files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]

    # Check if there are any images in the folder
    if not image_files:
        print(f"No images found in the folder '{folder_path}'.")
        return

    # Iterate through each image file
    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)
        try:
            # Open the image using PIL
            img = Image.open(image_path)

            # Crop the image using specified coordinates
            cropped_img = img.crop((left, top, right, bottom))

            # Save the cropped image, replacing the original
            cropped_img.save(image_path)
            print(f"Image '{image_file}' cropped and replaced successfully.")
        except Exception as e:
            print(f"Error cropping image '{image_file}': {str(e)}")


# Replace 'your_folder_path' with the actual path to your image folder in Google Colab
folder_path = '/content/Images'

# Replace these values with your desired cropping coordinates
left = 52
top = 44
right = 338
bottom = 312

# Crop and replace all images in the folder with specified coordinates
crop_and_replace_images(folder_path, left, top, right, bottom)


```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Conv2D, MaxPooling2D, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import plot_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.constraints import MinMaxNorm


# Load images and convert to NumPy array
image_path_prefix = '/content/Images/'  # Replace with the actual path to your images
image_size = (128, 128)  # Adjust the size based on your images

def load_and_preprocess_image(image_path):
    img = load_img(image_path, target_size=image_size)
    img_array = img_to_array(img)
    return img_array

df['image_path'] = image_path_prefix + df['Image']  # Replace with the actual column name
df['image_array'] = df['image_path'].apply(load_and_preprocess_image)

image_arrays = np.stack(df['image_array'].to_numpy())
target_columns = df[['T_by_R', 'L_by_R', 'B_by_R']].to_numpy()  # Replace with actual column names

# Normalize target columns if needed
# You may want to normalize these columns based on your specific requirements

# Clip values to be between 0 and 3
target_columns = np.clip(target_columns, 0, 3)

# Split the data into training and testing sets
X_train_img, X_test_img, y_train, y_test = train_test_split(
    image_arrays, target_columns, test_size=0.2, random_state=42)

def create_cnn_model():
    # Image input
    img_input = Input(shape=(image_size[0], image_size[1], 3), name='image_input')
    x = Conv2D(32, (3, 3), activation='relu')(img_input)
    x = MaxPooling2D((2, 2))(x)
    x = Conv2D(64, (3, 3), activation='relu')(x)
    x = MaxPooling2D((2, 2))(x)
    x = Conv2D(128, (3, 3), activation='relu')(x)
    x = MaxPooling2D((2, 2))(x)
    x = Flatten()(x)

    # Output layer with 3 neurons for each target column and linear activation function
    output = Dense(3, activation='linear', name='output')(x)

    # Create and compile the model
    model = Model(inputs=img_input, outputs=output)
    model.compile(optimizer=Adam(lr=0.01), loss='mean_squared_error', metrics=['mae'])

    return model

# Create the model
model = create_cnn_model()

# Display the model summary
model.summary()

# Train the model
model.fit(X_train_img, y_train, epochs=100, batch_size=8, validation_split=0.2)

# Evaluate the model
evaluation = model.evaluate(X_test_img, y_test)
print(f'\nTest Loss: {evaluation[0]}, Test MAE: {evaluation[1]}')
```

```python
import matplotlib.pyplot as plt

# Use the trained model to make predictions on the test dataset
predictions = model.predict(X_test_img)

# Choose a few random indices to visualize
random_indices = np.random.choice(len(X_test_img), size=5, replace=False)

# Visualize the results
for idx in random_indices:
    # Display the original image
    original_img_path = df.iloc[idx]['image_path']
    original_img = load_img(original_img_path)
    plt.imshow(original_img)
    plt.title(f"Original Image: {original_img_path}")
    plt.show()

    # Display the predicted values and the actual values
    actual_values = y_test[idx]
    predicted_values = predictions[idx]
    print(f"Actual Values: {actual_values}")
    print(f"Predicted Values: {predicted_values}")
    print("=" * 50)

```
